# Milestone 1: Introduction to LBE and Streaming Operator

```python
import numpy as np
from src.simulation import LatticeBoltzmann
```

## Parameters

```python
size_x = 20
size_y = 12
n_steps = 100
```

## Initialize lattice with PDF

```python
pdf = np.zeros((9, size_y, size_x))
# streaming directions/channels: east(1), south(4), south-east(8)
pdf[[1, 4, 8], :3, :3] = 1
lattice = LatticeBoltzmann(size_x, size_y, init_pdf=pdf)
```

## Run simulation steps

```python
for i in range(n_steps):
    lattice.update_density()
    lattice.update_velocity()
    lattice.streaming_step()
    lattice.update_plot()

anim = lattice.plot.display_animation()
```
